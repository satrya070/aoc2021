```python
mapp = {}
for line in listo[2:]:
    idx, val = line.split(' -> ')
    mapp[idx]  = val
```

```python
def run(inp):
    used_pairs = set()
    output = ''
    total_pairs = len(inp) - 1
    
    p1 = 0
    p2 = 1
    
    for pair in range(total_pairs):
        pair_idx = inp[p1] + inp[p2]
        value = mapp[pair_idx]
        used_pairs.add(pair_idx)
        
        output += inp[p1] + value
        
        p1 += 1
        p2 += 1
    
    output += inp[-1]
    
    print(used_pairs)
    
    return output
```

```python
run('NNCB')
```

```python
def countOccurrences(output):
    counts = {}

    for char in output:
        if char not in counts:
            counts[char] = {}
            counts[char]['cnt'] = 0
        counts[char]['cnt'] += 1
    
    for char in counts:
        counts[char]['perc'] = counts[char]['cnt'] / len(output)
        
    return counts
```
